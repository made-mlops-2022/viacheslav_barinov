```python
THRESHOLD = 0.7 # гиперпараметр решения

strong_corr = []
for name, row in data.corr().iterrows():
  for i, j  in enumerate(row):
    if abs(j) > THRESHOLD and data.columns[i] != name\
     and (data.columns[i], name, j) not in strong_corr:
      strong_corr.append((name, data.columns[i], j))
print(strong_corr)
```
